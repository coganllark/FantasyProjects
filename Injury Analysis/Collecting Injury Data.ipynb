```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
cols = ["Player", "Team", "Position", "Injuries", "Practice Status", "Game Status", "Game Time", "Week"]
injury_df = pd.DataFrame(columns = cols)

weeks = list()
week_names = list()
for i in range(2008, 2021):
    for j in range(1, 18):
        weeks.append(str(i) + "/REG" + str(j))
        week_names.append("Week " + str(j) + ", " + str(i))
```

```python
#221 weeks
for week_num in range(len(weeks)):
    week = weeks[week_num]
    week_name = week_names[week_num]
    print(week_name)
    # For each week request that week's injury report
    url = "https://www.nfl.com/injuries/league/" + week
    page = requests.get(url)
    print(page)
    soup = BeautifulSoup(page.text, 'lxml')
    
    # Get the teams and times from that week
    teams = soup.find_all("span", class_ ="nfl-c-matchup-strip__team-abbreviation")
    teams = [i.text for i in teams]
    teams = [i.strip() for i in teams]

    times = soup.find_all("span", class_ ="nfl-c-matchup-strip__date-time")
    times = [i.text for i in times]
    times = [i.strip() for i in times]
    
    team_counter = 0
    time_counter = 0
    # For each table on webpage
    for table in soup.find_all('table'):
        # For each tr in table
        team = teams[team_counter]
        time = times[int(time_counter/2)]
        team_counter += 1
        time_counter += 1
        for tr in table.find_all('tr')[1:]:
            # Get row data and put in df
            row_data = tr.find_all('td')
            row = [i.text for i in row_data]
            row = [i.strip() for i in row]
            row.insert(1, team)
            row.append(time)
            row.append(week_name)
            length = len(injury_df)
            injury_df.loc[length] = row
```

```python
injury_df.to_csv("NFL_Injury_Reports_2008_to_")
```
