```python
import pandas as pd
```

```python
full_data = pd.DataFrame(columns=["Player", "Pos", "Tm", "Week", "PPRFantasyPoints", "PPG"])
```

```python
ppg_data = pd.read_csv("PPG_2008_to_2020.csv")
ppg_data
```

```python
weeks = list()
week_names = list()
for i in range(2008, 2021):
    for j in range(1, 18):
        weeks.append(str(i) + "/week" + str(j) + ".csv")
        week_names.append("Week " + str(j) + ", " + str(i))
```

```python
# Building scoring data frame with PPG
url = "https://raw.githubusercontent.com/fantasydatapros/data/master/weekly/"
for i in range(len(weeks)):
    week_name = week_names[i]
    year = int(week_name[-4:])
    curr_url = url + weeks[i]
    curr_data = pd.read_csv(curr_url)
    
    week_name_list = list()
    for j in range(len(curr_data)):
        week_name_list.append(week_name)
    
    ppg_list = list()
    np_data = curr_data.to_numpy()
    for row in np_data:
        curr_name = row[0]
        curr_ppg = -1
        player_frame = ppg_data.loc[(ppg_data['Player'] == curr_name) & (ppg_data['Year'] == year)]
        if len(player_frame) > 0:
            curr_ppg = ((ppg_data.loc[(ppg_data['Player'] == curr_name) & (ppg_data['Year'] == year)])['PPG']).to_numpy()[0]
        ppg_list.append(curr_ppg)
    
    curr_data['Week'] = week_name_list
    curr_data['PPG'] = ppg_list
    full_data = full_data.append(curr_data[["Player", "Pos", "Tm", "Week", "PPRFantasyPoints", "PPG"]])
```
