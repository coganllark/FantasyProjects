# Appends Year 1 Receiving Stats from PFR to Pahowdy CSV

```python
import pandas as pd
import numpy as np
import requests
import responses
from bs4 import BeautifulSoup
```

```python
def get_stats(pfr_id):
    """
    Inputs:
    pfr_id -- Receivers Pro Football Reference ID (str)
    Returns:
    stats -- Tuple beginning with receivers yards, followed by yards per game (tup)
    """
    url = "https://www.pro-football-reference.com/players/" + str(pfr_id)[0] + "/" + str(pfr_id) + ".htm"
    response = requests.get(url)
    
    stats = (None, None)
    
    print(url)
    
    # If url with that user was found
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')
        table = soup.find("table", {"id":"receiving_and_rushing"})
        # If reciever has NFL Stats
        if table != None:
            body = table.find("tbody")
            first = body.find("tr")
            row = first.find_all("td")
            yards = float(row[8].text)
            yards_per = float(row[14].text)
            stats = (yards, yards_per)
    
    return stats
```

```python
data = pd.read_csv("cfb_data_1_14_22_WR.csv", skiprows=2)
cols = list(data.columns)
cols.append("Year 1 Yards")
cols.append("Year 1 Yards/G")
```

```python
cols
```

```python
data = data.to_numpy()
```

```python
new_data = list()
```

```python
for i in range(len(data)):
    pfr_id = data[i][2]
    
    stats = get_stats(pfr_id)
    
    new = np.append(data[i], stats[0])
    new = np.append(new, stats[1])
    new_data.append(new)
```

```python
new_frame = pd.DataFrame(new_data, columns=cols)
```

```python
new_frame.to_csv("cfb_data_1_14_22_WR_year1.csv")
```
