```python
import pandas as pd
import numpy as np
import requests
import responses
from bs4 import BeautifulSoup
```

```python
url = "https://www.playerprofiler.com/nfl/davante-adams/"
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
advanced = soup.find("section", {"id":"advanced-metrics"})
```

```python
adv_met_tabs = soup.find("advanced-metrics-tabs")
```

```python
adv_str = str(adv_met_tabs)
```

```python
# Returns a list of years options for advanced metrics dropdown
for i in range(len(adv_str)):
    if adv_str[i:i+5] == "years":
        years = list()
        j = i+8
        while adv_str[j] != "]":
            possible_year = adv_str[j:j+4]
            if possible_year.isnumeric():
                years.append(int(possible_year))
            j += 1
        break
```
